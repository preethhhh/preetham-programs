```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load data and select relevant columns
data = pd.read_csv("Mall_Customers.csv")
X = data.iloc[:, [3, 4]].values

# Calculate inertia for each number of clusters (1-10) and plot the elbow graph
inertia = [KMeans(n_clusters=i, init='k-means++', random_state=42).fit(X).inertia_ for i in range(1, 11)]
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()

# Apply KMeans with optimal clusters (5 in this example) and plot results
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualize clusters
for i, color in enumerate(['blue', 'green', 'lime', 'red', 'black']):
    plt.scatter(X[y_kmeans == i, 0], X[y_kmeans == i, 1], s=100, color=color, label=f'Cluster {i+1}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, color='yellow', label='Centroids')
plt.title("Customer Clusters")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```
